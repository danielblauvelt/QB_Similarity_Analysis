```python
import os
import gensim
import pandas as pd
```

```python
docs = pd.read_csv('all_draft_classes.csv')
docs['Strengths'] = 'STRENGTHS: ' + docs['Strengths'].astype(str)
docs['Weaknesses'] = 'WEAKNESSES: ' + docs['Weaknesses'].astype(str)

docs['composite'] = docs['Overview'] + docs['Strengths'] + docs['Weaknesses']
```

```python
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
stop_words = stopwords.words('english')
lemmatizer = WordNetLemmatizer()
# Tokenize the sentence
overviews = []
for doc in docs['composite']:
    # tokens = word_tokens = word_tokenize(doc)
    filtered_sentence = [w for w in doc.split() if not w.lower() in stop_words]
    filtered_sentence = [lemmatizer.lemmatize(word) for word in filtered_sentence]
    overviews.append(" ".join(filtered_sentence).lower())
```

```python
from gensim.models.doc2vec import Doc2Vec, TaggedDocument
#Create a list of tagged documents. Documents need to be tagged for other doc2vec methods to work.
documents = [TaggedDocument(doc.split(), [i]) for i, doc in enumerate(overviews)]
```

```python
max_epochs = 100
vec_size = 20
alpha = 0.025

model = Doc2Vec(vector_size=vec_size,
                alpha=alpha, 
                min_alpha=0.00001)
  
model.build_vocab(documents)

for epoch in range(max_epochs):
    print('iteration {0}'.format(epoch))
    model.train(documents,
                total_examples=model.corpus_count, 
                epochs = model.epochs)
    # decrease the learning rate
    model.alpha -= 0.0002
    model.min_alpha = 0.00001
```

```python
from gensim.test.utils import get_tmpfile
fname = get_tmpfile("my_doc2vec_model")
model.save(fname)
model = Doc2Vec.load(fname)
```

```python
#Create player comparisons
for i in range(len(docs)):
    player = model.docvecs.most_similar([i], topn=3)
    print('Player', docs.iloc[i]['Name'], 'is similar to')
    for p in player:
        print(p[0], docs.iloc[p[0]]['Name'], 'with similarity', p[1])
```
